```python
f = open("chave_consumidor.txt", "r")
chave_consumidor = f.read()
f = open("segredo_consumidor.txt", "r")
segredo_consumidor = f.read()
f = open("token_acesso.txt", "r")
token_acesso = f.read()
f = open("token_acesso_segredo.txt", "r")
token_acesso_segredo = f.read()
```

```python
autenticacao = tweepy.OAuthHandler(chave_consumidor, segredo_consumidor)
```

```python
autenticacao = tweepy.OAuthHandler(chave_consumidor, segredo_consumidor)
autenticacao.set_access_token(token_acesso, token_acesso_segredo)
```

```python
twitter = tweepy.API(autenticacao)
```
